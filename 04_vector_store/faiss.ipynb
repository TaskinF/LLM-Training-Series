### FAISS

- FAISS (Facebook AI Similarity Search) is a fast and efficient vector search library developed by Facebook.

    - It finds close neighbors (similar vectors) in a large number of vectors.

    - It has CPU and GPU support.

    - It is especially widely used in embedding-based search/RAG systems.

```python
from langchain_community.document_loaders import TextLoader
from langchain_community.vectorstores import FAISS
from langchain_community.embeddings import OllamaEmbeddings
from langchain_text_splitters import CharacterTextSplitter
```

```python
loader = TextLoader('speech.txt')
documents = loader.load()

text_splitter = CharacterTextSplitter(chunk_size = 1000, chunk_overlap = 30)
docs = text_splitter.split_documents(documents)
```

```python
docs
```

```python
embeddings = (
    OllamaEmbeddings(model="nomic-embed-text") # A high-performing open embedding model with a large token context window. 
)

db = FAISS.from_documents(docs, embeddings)
db
```

```python
# Query
query = "How does the speaker describe the desired outcome of the war?"
docs = db.similarity_search(query)
docs[0].page_content
```

#### Retriever

```python
retriever = db.as_retriever()
docs = retriever.invoke(query)
docs[0].page_content
```

#### similarity_search_with_score

- It brings not only similar documents but also a similarity score.
- Returned distance score is L2 distance. Therfore, a lower score is better

- What does it do?

    - It shows how many points match (for example 0.85).

    - This way we can set a confidence threshold (like score > 0.75), we can filter out very low matches.




```python
docs_and_score = db.similarity_search_with_score(query)
docs_and_score
```

```python
embedding_vector = embeddings.embed_query(query)
embedding_vector
```

```python
docs_score = db.similarity_search_by_vector(embedding_vector)
docs_score
```

```python
## Saving and Loading
db.save_local("faiss_index")
```

```python
new_db = FAISS.load_local("faiss_index", embeddings, allow_dangerous_deserialization=True)
docs = new_db.similarity_search_with_score(query)
```
